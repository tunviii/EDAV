# 4. Reindexing
# Reindexing changes the row or column labels of a Series or DataFrame.


```python
#df.shape
#df.info()
#df.describe()
#df.head(3)
#df.tail(3)
#df.iloc[1:4, :3] #rows, columns
#df.loc[1:3, ["Name","Department"]]  #label based indexing
#df.drop("Age", axis = 1)
df.rename(columns = {"Department": "Dept"}, inplace = True)
```

```python
#df["Dept"].unique()
df["Dept"].value_counts()
```

```python
df.isnull().sum()
df.dropna(how = "any") # any row that had any null value
df.dropna(how = "all") #if all the values in any row are null then we drop that row
df["Age"].fillna(df["Age"].mean())
df["Salary"].fillna(df["Salary"].median())
df["Age"].fillna(method = "ffill")
df["Age"].fillna(method = "bfill")
df["Name"] = df["Name"].replace("Charlie", "Rose")
```

```python
# Joins and Merges

department_info = {
    "Dept": ["HR", "IT", "Finance"],
    "Location": ["New York", "San Francisco", "Chicago"],
    "Manager": ["Laura", "Steve", "Nina"]
}

df2 = pd.DataFrame(department_info)
df2

# Concat
pd.concat([df, df2])
pd.concat([df, df2], axis = 1)

#merge
pd.merge(df, df2, on = "Dept")
```

```python
import pandas as pd
print("\n# Reindexing a Series")
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = s.reindex(['c', 'a', 'b', 'd'])
print(s)
print(s2)
```

```python
df3.reindex(columns=['A', 'B', 'C'])
```

```python

print("\n# Reindexing with forward fill")
s3 = pd.Series(['red', 'blue', 'green'], index=[0, 2, 4])
print(s3.reindex(range(10), method='ffill'))
```

```python
print("\n# Reindexing with forward fill using DataFrame")
df3 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Paris', 'London']
})
print(df3.reindex(range(6), method='ffill'))
```

# 5. Dropping Entries
# Use drop() to remove rows or columns from a Series or DataFrame.

```python
print("\n# Dropping from Series")
series3 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
# print(series3)
print(series3.drop('b'))
```

```python
print("\n# Dropping from DataFrame")
df3 = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Paris', 'London']
}
df4 = pd.DataFrame(df3)
print(df4)
# print(df4.drop(1))
print(df4.drop('Age', axis=1)) # removing column with name Age

```

```python

# Dropping the row where Name is 'Alice'
# First, we need to find the index of the row where Name == 'Alice'
row_index = df4[df4['Name'] == 'Alice'].index
# row_index
print("\nDataFrame after dropping row with Name = 'Alice':")
print(df4.drop(row_index))
```
